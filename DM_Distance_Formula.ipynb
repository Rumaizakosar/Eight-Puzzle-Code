```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
```

```python
from google.colab import files
uploaded = files.upload()

titanic_data = pd.read_csv('/content/Titanic-Dataset.csv')
titanic_data.head()
```

```python
titanic_data = titanic_data.dropna()  # Example of dropping missing values
titanic_data['Sex'] = titanic_data['Sex'].map({'male': 1, 'female': 0})  # Encode categorical variable
```

```python
numerical_data = titanic_data[['Age', 'Fare']]  # Select numerical columns
euclidean_distances = pairwise_distances(numerical_data, metric='euclidean')
manhattan_distances = pairwise_distances(numerical_data, metric='manhattan')

print("Euclidean Distances:\n", euclidean_distances)
print("Manhattan Distances:\n", manhattan_distances)
```

```python
categorical_data = titanic_data[['Pclass', 'Sex', 'Embarked']]  # Select categorical columns
categorical_data = pd.get_dummies(categorical_data)  # Convert categorical to numerical

hamming_distances = pairwise_distances(categorical_data, metric='hamming')

print("Hamming Distances:\n", hamming_distances)

```
